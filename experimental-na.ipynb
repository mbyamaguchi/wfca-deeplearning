```python
import csv
import numpy as np

resultpath = "../neuralactivities/results_2400.csv"

result = []

with open(resultpath, newline='') as csvfile:
    line = csv.reader(csvfile, delimiter=',')
    for row in line:
        result.append(row)
    
result = np.array(result)
```

```python
import pandas as pd

df = pd.read_csv(resultpath)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(3, 1)

x = np.arange(0, 2400, 1)

for i, r in enumerate(result.T[:3]):
    ax[i].plot(x, r)
```
